### Itens 4 a 6 - Resolução:

```python
dados.packpc.plot(kind='hist', legend= 'False',bins=200,normed=True)
```

## O histograma acima pode ser comparado à duas distribuições teóricas: norm e gamma. A distribuição "norm" cresce até o pico no centro e depois decresce. Já a "gamma" se assemelha à norm, porém deslocada um pouco para a esquerda e com decréscimo maior, ou seja, o pico se encontra mais a esquerda.

## Além disso, o gráfico possui uma distribuição simétrica pois a média, a moda e a mediana possuem quase o mesmo valor. Isso ocorre pois os dados estão concentrados no centro, característica da distribuição normal.

```python
media = dados.packpc.mean()
print(media)

fit_norm= stats.norm.fit(dados.packpc)
fit_gamma = stats.gamma.fit(dados.packpc)
print(fit_norm)
print(fit_gamma)
```

## O resultado da função fit() retornou valores que fazem sentido para 'norm' porém não para gamma. Se analisados, os valores de média, por exemplo, divergem entre o histograma e a função.

### Itens 7 a 9 - Resolução:

```python
x = np.linspace(-100, 300, 200)
y = norm.pdf(x, loc=100, scale=20)
plt.plot(x,y, color = "r")


g= np.linspace(-100,300,100)
k = 20
media = dados.packpc.mean()
theta = media/k
z= stats.gamma.pdf(g, scale=theta, loc=8, a=k)


plt.plot(g, z, '-', color='yellow', label="Chi")
dados.packpc.plot(kind='hist', legend= 'False',bins=20,normed=True)
```
